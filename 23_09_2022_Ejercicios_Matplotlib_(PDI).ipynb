```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x = ['enero', 'febrero', 'marzo', 'abril', 'mayo', 'junio', 'julio', 'agosto', 'septiembre', 'octubre', 'noviembre', 'diciembre']
```

```python
y= [13500, 13300, 14000, 13900, 15100, 14560, 15000, 15300, 17000, 18300, 20000, 21300]
```

```python
plt.figure (figsize = (20, 6))
plt.title ('Gastos del supermercado')
plt.xlabel ('meses')
plt.ylabel ('gasto')
plt.plot (x, y)
plt.show()
```

```python
plt.figure (figsize = (20, 6))
plt.title ('Gastos del supermercado')
plt.xlabel ('meses')
plt.ylabel ('gasto')
plt.bar (x, y)
plt.show()
```
